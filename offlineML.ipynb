# volume symbol lowPrice highPrice closeTime count openPrice openTime lastPrice

```python
from pycaret.regression import *
import pandas as pd
data = pd.read_csv("btc_usdt.csv",names=["volume","symbol","lowPrice","highPrice","closeTime","count","openPrice", "openTime","lastPrice"])
#print(data.head())
print(data.shape)
```

```python
s = setup(data, target = 'lastPrice', session_id = 123,ignore_features=["symbol","closeTime","openTime"],index="closeTime")
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
firstModel = create_model('et')
```

```python
save_model(firstModel, 'model_et')
```

```python
secondModel = create_model('rf')
```

```python
save_model(secondModel, 'model_rf')
```
